Use [earthkit-plots](https://earthkit-plots.readthedocs.io/) to make visualizations of temperature data stored in a NetCDF file. 

```python
import earthkit
```

We have downloaded a NetCDF file from the "[ERA5-Land monthly averaged data from 1950 to present](https://cds.climate.copernicus.eu/datasets/reanalysis-era5-land-monthly-means)" dataset on the [Climate Data Store](https://cds.climate.copernicus.eu/). The file contains the mean temperature at 2m for February 2025 and we can load it with [earthkit-data](https://earthkit-data.readthedocs.io/): 

```python
file = "../data/era5-land-temperature-2m-monthly-mean-feb-2025.nc"
temperature = earthkit.data.from_source("file", file)
```

The quickest way to get started with earthkit-plots is with the [quickplot](https://earthkit-plots.readthedocs.io/en/latest/examples/guide/02-quickplot.html) function, which will take your data and attempt to generate something appropriate:

```python
earthkit.plots.quickplot(temperature)
```

We see a global map with temperatures is in kelvin. If we want to switch to celcius and focus on Africa we can add some parameters: 

```python
earthkit.plots.quickplot(temperature, units="celsius", domain="Africa")
```

The Map class gives you more control provides various methods for plotting and enhancing visualisations. Below is an example of how to create a map, add title and legend, and overlay coastlines, borders and cities.

```python
map = earthkit.plots.Map(domain="Ethiopia")
map.quickplot(temperature, units="celsius")
map.legend(label="{variable_name} ({units})")
map.title("ERA5-Land mean {variable_name} in {time:%B %Y}")
map.coastlines()
map.borders()
map.cities()
map.show()
```
