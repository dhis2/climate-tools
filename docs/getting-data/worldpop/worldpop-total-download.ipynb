In this notebook we will use `dhis2eo` to download [yearly population count data from WorldPop](https://hub.worldpop.org/geodata/listing?id=135) for a given country. 

-------------
## What you need

Start by importing the libraries that we need:

```python
import dhis2eo
import dhis2eo.org_units
from dhis2eo.data.worldpop import population
```

In the case of WorldPop, data downloads are organized by country, so the only other information we need is the ISO3 code of the country you want to download data for. 

## Downloading WorldPop total population data

The `dhis2eo.data.worldpop.population` module allows you do easily download yearly historical and projected population data for individual countries from the [WorldPop Global v2 dataset between 2015 and 2030](https://hub.worldpop.org/project/categories?id=3). Downloading population data for a country and range of years can be achieved with a simple yearly loop and the country's ISO3-code: 

```python
country_code = 'SLE'
start_year = 2015
end_year = 2030
for year in range(start_year, end_year+1):
    print(f'Year: {year}')

    # Download data...
    data = population.get(year, country_code)

    # do something with the data
    # e.g. save to disk, aggregate, or import to DHIS2
    # ...
```

To inspect the contents of the downloaded data, let's view the data of the last element of the loop:

```python
data
```

We see that this is data for the last available year, 2030, and contains data variables `total_pop` (Total population).
