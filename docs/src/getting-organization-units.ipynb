Downloading organization units from the DHIS2 Web API

```python
import geopandas
import io
import json
from dhis2_client import DHIS2Client
from dhis2_client.settings import ClientSettings
```

```python
cfg = ClientSettings(base_url='http://localhost:8080', username='admin', password='district')
```

```python
client = DHIS2Client(settings=cfg)
```

```python
org_units_dict = client.get("/api/organisationUnits.geojson", params={"level": 2})
# print(org_units)
```

```python
org_units_string = json.dumps(org_units_dict)

gdf = geopandas.read_file(io.StringIO(org_units_string))
gdf.plot()
```
